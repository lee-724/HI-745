# Useful RDD Transformation Functions

Here are important RDD transformation and action functions used in Spark when working with RDDs:

---

## 1. `rdd.map(function)`
- Applies a function to each element of the RDD.
- Similar to Python's `map()`, it returns a new RDD.

**Example:**
```python
rdd = sc.parallelize([1, 2, 3, 4])
result = rdd.map(lambda x: x * 2).collect()
print(result)  # Output: [2, 4, 6, 8]



## 2. `rdd.flatMap(function)`
- Similar to `map()`, but it "flattens" the output.
- If the function returns a list or collection, `flatMap()` expands it into individual elements instead of keeping the nested structure.

**Example:**
```python
rdd = sc.parallelize(["hello world", "spark rdd"])
result = rdd.flatMap(lambda x: x.split()).collect()
print(result)  # Output: ['hello', 'world', 'spark', 'rdd']


Use Case:
When you need to break down elements into smaller parts. For example, splitting lines of text into individual words.

## 3. `rdd.filter(function)`
- Filters elements from the RDD based on a function that returns a Boolean (`True` or `False`).
- Only elements that satisfy the condition (`True`) are included in the resulting RDD.

**Example:**
```python
rdd = sc.parallelize([1, 2, 3, 4, 5])
result = rdd.filter(lambda x: x % 2 == 0).collect()
print(result)  # Output: [2, 4]


Use Case:
When you need to remove unwanted elements from an RDD, such as filtering out records that do not meet a specific condition.

## 4. `rdd.reduceByKey(function)`
- Combines values with the same key using a specified function.
- It performs the reduction in parallel, which is efficient for large datasets.

**Example:**
```python
rdd = sc.parallelize([("a", 2), ("a", 3), ("b", 5)])
result = rdd.reduceByKey(lambda x, y: x + y).collect()
print(result)  # Output: [('a', 5), ('b', 5)]



Use Case:
When you need to perform aggregations such as summing, finding the maximum, or any other operation for values grouped by key.

## 5. `rdd.groupByKey()`
- Groups all values with the same key into a single collection (e.g., a list or iterator).
- Unlike `reduceByKey()`, this does not perform any aggregation; it only groups the values.

**Example:**
```python
rdd = sc.parallelize([("a", 1), ("b", 2), ("a", 3)])
result = rdd.groupByKey().mapValues(list).collect()
print(result)  # Output: [('a', [1, 3]), ('b', [2])]


Use Case:
When you need to access all values associated with each key, for example, to calculate more complex metrics like averages or custom statistics.

## 6. `rdd.sortByKey()`
- Sorts the RDD by key values in either ascending (default) or descending order.
- It returns a new RDD where elements are sorted based on their keys.

**Example:**
```python
rdd = sc.parallelize([(2, "b"), (1, "a"), (3, "c")])
result = rdd.sortByKey().collect()
print(result)  # Output: [(1, 'a'), (2, 'b'), (3, 'c')]


Use Case:
When you need to order data by keys, which is often required for reports, analytics, or preparing data for further ordered processing.

## 7. `rdd.keys()`, `rdd.values()`
- `rdd.keys()`: Returns an RDD consisting of only the keys from a key-value RDD.
- `rdd.values()`: Returns an RDD consisting of only the values from a key-value RDD.

**Example:**
```python
rdd = sc.parallelize([("a", 1), ("b", 2), ("c", 3)])
keys = rdd.keys().collect()
values = rdd.values().collect()

print("Keys:", keys)      # Output: Keys: ['a', 'b', 'c']
print("Values:", values)  # Output: Values: [1, 2, 3]


Use Case:
When you only need to extract and work with keys or values separately in a key-value RDD.

## 8. `rdd.mapValues(function)`
- Applies a function to the values of each key-value pair in an RDD while keeping the keys unchanged.
- This is more efficient than using `map()` when you only need to transform values.

**Example:**
```python
rdd = sc.parallelize([("a", 2), ("b", 3)])
result = rdd.mapValues(lambda x: x * 10).collect()
print(result)  # Output: [('a', 20), ('b', 30)]


Use Case:
When you want to modify only the values in a key-value RDD while preserving the keys, e.g., scaling or applying a transformation to numeric values.

## 9. `rdd.collect()`
- Triggers the execution of all transformations on the RDD and returns the results as a list to the driver program.
- Use this action carefully, especially with large datasets, as it brings all data into memory on the driver.

**Example:**
```python
rdd = sc.parallelize([1, 2, 3, 4])
result = rdd.collect()
print(result)  # Output: [1, 2, 3, 4]


Use Case:
When you want to retrieve and print the final results from an RDD. Useful for small datasets or debugging but should be avoided with large RDDs due to memory constraints.

## Summary of String Operations and Methods

1. **Concatenation (`+`)**: Combines two strings.  
   Example: `'Hello' + ' World!'` → `'Hello World!'`

2. **Repetition (`*`)**: Repeats the string multiple times.  
   Example: `'Hello ' * 2` → `'Hello Hello '`

3. **Indexing and Slicing (`[]`, `[:]`)**: Accesses specific characters or substrings.  
   Example: `'Hello'[0:4]` → `'Hell'`

4. **Membership Operators (`in`, `not in`)**: Checks if a substring is present in a string.  
   Example: `'Hello' in 'Hello World'` → `True`

5. **Methods**:
   - **`capitalize()`**: Capitalizes the first letter of the string.  
     Example: `'hello'.capitalize()` → `'Hello'`
   - **`lower()`**, **`upper()`**: Converts all characters to lowercase or uppercase.  
     Example: `'HELLO'.lower()` → `'hello'`, `'hello'.upper()` → `'HELLO'`
   - **`encode()`**: Encodes the string in the specified format (e.g., UTF-8).  
     Example: `'Hello'.encode("utf-8")` → `b'Hello'`
   - **`split()`**: Splits the string by a delimiter.  
     Example: `'Hello World!'.split(" ")` → `['Hello', 'World!']`

6. **String Escaping**: Handles special characters such as newlines (`\n`) and tabs (`\t`).  
   Example: `print("This is a line.\nThis is another line.")`  

7. **Casting to String (`str()`)**: Converts other data types to strings.  
Example: `str(123)` → `'123'`

8. **Casting to Integer (`int()`)**: Converts a string containing numeric characters to an integer.  
Example: `int('123')` → `123`


## 1. Dictionary

A **dictionary** is a data structure that stores key-value pairs. Each key is unique and is used to access its corresponding value. Dictionaries are created using curly braces `{}`. Keys and values are separated by colons `:` and key-value pairs are separated by commas `,`.

### **Creating a Dictionary**
A dictionary is created by specifying key-value pairs inside curly braces. You can add new pairs to a dictionary after its creation.

### **Accessing Values**
To access a value, use a key inside square brackets `[]`. The key must exist in the dictionary; otherwise, a `KeyError` will be raised.

### **Key Points**
1. **Keys must be unique.** Duplicate keys are not allowed.
2. You can **add, access, and update** key-value pairs.
3. The **key** is used as an index to retrieve the corresponding value.

---

```python

## 2. List

A **list** is a data structure that stores an ordered sequence of objects. Lists can contain elements of any data type, including other lists and dictionaries. Items in a list are separated by commas `,` and enclosed within square brackets `[]`.

### **Creating a List**
A list is created by specifying a sequence of elements inside square brackets. Each element in the list is separated by a comma.

### **Accessing List Elements**
List elements are accessed by their index, which starts at 0. You can use the index inside square brackets `[]` to retrieve or modify elements.

### **Key Points**
1. **Lists are ordered**: The order of elements in a list is maintained.
2. **Indexing starts from 0**: The first element is accessed with `list[0]`, the second with `list[1]`, and so on.
3. Lists can **store multiple data types**, including strings, numbers, and other lists or dictionaries.

# Word Count Improvement (2) – Regular Expression

---

## **Problem**

- The word count program may encounter variations in words due to different capitalization and punctuation.
  - For example, `"Hello"`, `"hello,"`, and `"HELLO"` should all be treated as the same word.
- This inconsistency can lead to incorrect word counts.

---

## **Solution**

- We will use **regular expressions** to:
  1. Normalize words by converting them to lowercase.
  2. Split text by non-word characters (e.g., punctuation, whitespace).
  
- Python's **`re`** module provides support for regular expressions.

### **Python Regex Documentation**
[Python Regular Expression Documentation](https://docs.python.org/2/library/re.html)

---

## **Regular Expression Approach**

- We define a function to normalize words using regular expressions:
  ```python
  import re

  def normalizeWords(text):
      # Convert text to lowercase and split by non-word characters
      return re.compile(r'\W+', re.UNICODE).split(text.lower())
      

````python
# Import regular expressions (re) module and PySpark modules
import re


# Step 1: Define a function to normalize words
# The function converts text to lowercase and splits it using a regex pattern that matches non-word characters
def normalizeWords(text):
    return re.compile(r'\W+', re.UNICODE).split(text.lower())

# Step 2: Set up the Spark configuration and context
conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext(conf=conf)

# Step 3: Read the input text file into an RDD
input = sc.textFile("2-book.txt")

# Step 4: Normalize and split words using the defined function
words = input.flatMap(normalizeWords)

# Step 5: Count the occurrences of each normalized word
wordCounts = words.countByValue()

# Step 6: Clean and print each word and its count
for word, count in wordCounts.items():
    # Encode the word to remove non-ASCII characters
    cleanWord = word.encode('ascii', 'ignore')

    # Only print words that have valid ASCII characters
    if cleanWord:
        print(cleanWord.decode() + " " + str(count))

sc.stop()
````

# Find the Most Popular Movie

In this exercise, we will determine which movie is watched most often using the **Movielens** dataset. The data is located in the `ml-100k` folder and the file is named `u.data`.

---

## **Understanding the Dataset**

The `u.data` file contains movie rating information. Each line in the file is structured as follows:

1. **User ID**: The ID of the user who watched and rated the movie.
2. **Movie ID**: The ID of the movie.
3. **Rating**: The user's rating for the movie (on a scale of 1 to 5).
4. **Timestamp**: The time when the rating was made.

### **Example Line in `u.data`**
196    242    3    881250949

- **User ID**: 196
- **Movie ID**: 242
- **Rating**: 3
- **Timestamp**: 881250949

---

## **Task**

We need to figure out which movie was watched most often by counting how many times each movie ID appears in the dataset.

---

## **Steps to Solve the Task**

1. Read the input data (`u.data`).
2. Extract the **movie ID** from each line.
3. Convert the movie ID to a key-value pair `(movie ID, 1)`.
4. Count the occurrences of each movie ID using `reduceByKey()`.
5. Sort the results by the count in descending order.
6. Print the most popular movie ID and its count.

---

## **Full Code**

```python
from pyspark import SparkConf, SparkContext

# Step 1: Set up Spark configuration and context
conf = SparkConf().setMaster("local").setAppName("MostPopularMovie")
sc = SparkContext(conf=conf)

# Step 2: Read the input file
input = sc.textFile("u.data")

# Step 3: Extract movie IDs
# We split each line and extract the second field (movie ID)
movieIDs = input.map(lambda line: line.split()[1])

# Step 4: Convert movie IDs to key-value pairs and count them
movieCounts = movieIDs.map(lambda movieID: (movieID, 1)).reduceByKey(lambda x, y: x + y)

# Step 5: Sort by the count in descending order
sortedMovies = movieCounts.map(lambda x: (x[1], x[0])).sortByKey(ascending=False)

# Step 6: Collect and print the results
for count, movieID in sortedMovies.collect():
    print(f"Movie ID {movieID}: {count} views")

# Advanced Technique: Broadcast Variables

Broadcast variables allow you to send large read-only data efficiently to all executors. Instead of repeatedly sending the data with each task, Spark transfers it once and stores it on each executor. This reduces data transfer overhead and improves performance for tasks that require the same reference data (e.g., movie names).

---

## **Key Points**

1. **Broadcast Mechanism**:
   - Broadcast variables are sent to all executors and remain available throughout the Spark job.
   - You can use `SparkContext.broadcast()` to create a broadcast variable.
   - The broadcast variable is immutable and accessible across all tasks on the executors.

2. **Usage**:
   - Create the broadcast variable using:
     ```python
     var1 = SparkContextObject.broadcast(variables)
     ```
   - To access the original object inside tasks, use `var1.value`.

3. **Limitations**:
   - Avoid broadcasting very large datasets, as it may lead to memory issues on the executors.

---

## **MovieLens Application Example**

In this example, we improve the **popular movies** application by displaying movie names instead of movie IDs. We use the `u.item` file, which contains movie ID-to-name mappings. This data is broadcast to all executors.

## **Steps**

2. Broadcast the movie dictionary to all executors.

conf = SparkConf().setMaster("local").setAppName("PopularMoviesWithNames")

````python
# Step 1: Configure and initialize the Spark context
conf = SparkConf().setMaster("local").setAppName("PopularMoviesWithNames")
sc = SparkContext(conf=conf)

# Step 2: Read the movie names from u.item and create a dictionary
def loadMovieNames():
    movieNames = {}
    with open("u.item", encoding="ISO-8859-1") as f:
        for line in f:
            fields = line.split('|')
            movieNames[int(fields[0])] = fields[1]
    return movieNames

# Step 3: Broadcast the movie names dictionary to executors
movieNamesBroadcast = sc.broadcast(loadMovieNames())

# Step 4: Read the input data file and extract movie IDs
lines = sc.textFile("u.data")
movies = lines.map(lambda x: (int(x.split()[1]), 1))

# Step 5: Count the occurrences of each movie ID
movieCounts = movies.reduceByKey(lambda x, y: x + y)

# Step 6: Use the broadcast variable to map movie IDs to movie names
mappedMovies = movieCounts.map(lambda x: (movieNamesBroadcast.value[x[0]], x[1]))

# Step 7: Sort by view count in descending order
sortedMovies = mappedMovies.sortBy(lambda x: x[1], ascending=False)

# Step 8: Collect and print the results
results = sortedMovies.collect()
for result in results:
    print(f"{result[0]}: {result[1]} views")

sc.stop()
````

````python
# Step 1: Configure and initialize Spark context
conf = SparkConf().setMaster("local").setAppName("PopularHero")
sc = SparkContext(conf=conf)

# Step 2: Define a function to count co-occurrences
def countCoOccurences(line):
    elements = line.split()
    return (int(elements[0]), len(elements) - 1)

# Step 3: Define a function to parse names from the file
def parseNames(line):
    fields = line.split('"')
    return (int(fields[0]), fields[1].encode("utf8"))

# Step 4: Read the superhero names data
names = sc.textFile("3-Marvel-Names.txt")
namesRdd = names.map(parseNames)

# Step 5: Read the superhero co-occurrence data
lines = sc.textFile("3-Marvel-Graph.txt")

# Step 6: Calculate the number of connections for each superhero
pairings = lines.map(countCoOccurences)
totalFriendsByCharacter = pairings.reduceByKey(lambda x, y: x + y)

# Step 7: Flip the (hero_id, connections) pair to (connections, hero_id)
flipped = totalFriendsByCharacter.map(lambda xy: (xy[1], xy[0]))

# Step 8: Find the superhero with the most connections
mostPopular = flipped.max()

# Step 9: Get the name of the most popular superhero
mostPopularName = namesRdd.lookup(mostPopular[1])[0]

# Step 10: Print the result
print(str(mostPopularName) + " is the most popular superhero, with " + \
      str(mostPopular[0]) + " co-appearances.")

sc.stop()
````
