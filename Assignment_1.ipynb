```python
# Import the needed tools to configure and control the system from pyspark library
from pyspark import SparkConf, SparkContext

# Create a 'Configuration' object to define how the program should run
# .setMaster("local") is commanding the code to run on your own computer
# .setAppName("Ratings_Histogram") gives the specific task a name for the logs
configuration = SparkConf().setMaster("local").setAppName("Ratings_Histogram")

# Creating the 'Spark Context', which is an entry point for all Spark functions
# It takes the configuration, and starts the engine
sc = SparkContext(conf = configuration)
print ('Apache Spark Started')
```

```python
# Import the 'requests' library
import requests

# Save the web address (URL) where the data exists in a vriable name for easy use
github_data_url = "https://raw.githubusercontent.com/Luo-Innovation-Lab/Big-Data/main/ml-100k/u.data"

# Download the data from the url that you are providing
response = requests.get(github_data_url)

# Check if the data were able to be downloaded
response.raise_for_status()

# Take the block of data and split it into a list of individual lines
data_lines = response.text.splitlines()

# Show only the first 5 lines to check if you got the correct data
data_lines[:5]
```

```python
# Take the list of lines and distribute them across the Spark system
# Turn the data into an 'RDD'
lines = sc.parallelize(data_lines)

# Mention the variable name to inspect the data
lines
```

```python
# Create a new data set called 'ratings' with lines named as x
# Cut the line into pieces wherever there is a 'Tab' space using '.split("\t")'
# Grab the ratings, since [2] means 3rd column of the data list
ratings = lines.map(lambda x: x.split("\t")[2])

# Count how many times each unique rating appears
# Return a dictionary with rating as 'key' and the count as 'value'
result = ratings.countByValue()

# Display the final counts
result
```

```python
# Import the 'collections' to help organizing the final data
import collections

# Take the final counts, sort them by the rating number
# Put them into an 'OrderedDict' in that specific order
sortedResults = collections.OrderedDict(sorted(result.items()))

# Loop through each pair (rating as 'key', count as 'value') in our sorted list
for key, value in sortedResults.items():
    # Print the result of the rating and how many people gave that ratings
    print(f"Rating {key}: {value} (Count)")
```
