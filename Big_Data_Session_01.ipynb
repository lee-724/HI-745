This is our first example of Apache Spark. Counting ratings of movies from users.

```python
print ("Let's get started")
```

```python
# From the pyspark library, import tools to configure and control the system
from pyspark import SparkConf, SparkContext # making edit

# Create a 'Configuration' object to define how the program should run
# .setMaster("local") tells the code to run on your own computer, not a giant cluster
# .setAppName("Ratings_Histogram") gives your specific task a name for the logs
configuration = SparkConf().setMaster("local").setAppName("Ratings_Histogram")

# Create the 'Spark Context', which is the "brain" or entry point for all Spark functions
# It takes the configuration we just made and starts the engine
sc = SparkContext(conf = configuration)
print ('Apache Spark Started')
```

```python
# Import the 'requests' library, which allows your code to talk to the internet
import requests

# Save the web address (URL) where the data lives into a variable name for easy use
github_data_url = "https://raw.githubusercontent.com/Luo-Innovation-Lab/Big-Data/main/ml-100k/u.data"

# Go to the URL and download the content; the result is stored in the 'response' variable
response = requests.get(github_data_url)

# Check if the download was successful (if the website is down, this will show an error)
response.raise_for_status()

# Take the giant block of text from the website and split it into a list of individual lines
data_lines = response.text.splitlines()

# Show only the first 5 lines of the data so we can check if it looks correct
data_lines[:5]
```

```python
# Take the list of lines (data_lines) and distribute them across the Spark system
# This turns the data into an 'RDD', which Spark can process much faster
lines = sc.parallelize(data_lines)

# Simply mention the variable name to 'inspect' it and see what kind of object it is
lines
```

```python
# Create a new set of data called 'ratings' by looking at every line (x)
# .split("\t") cuts the line into pieces wherever there is a 'Tab' space
# [2] grabs the 3rd piece of information (the rating), as computers start counting at 0
ratings = lines.map(lambda x: x.split("\t")[2])

# Count how many times each unique rating (like "1", "2", "3", etc.) appears
# This returns a dictionary where the rating is the 'key' and the count is the 'value'
result = ratings.countByValue()

# Display the final counts on the screen
result
```

```python
# Import the 'collections' toolset to help us organize our final data
import collections

# Take the 'result' counts, sort them by the rating number (1, 2, 3...),
# and put them into an 'OrderedDict' so they stay in that specific order
sortedResults = collections.OrderedDict(sorted(result.items()))

# Loop through each pair (key is the rating, value is the count) in our sorted list
for key, value in sortedResults.items():
    # Print a nicely formatted sentence showing the rating and how many people gave it
    print(f"Rating {key}: {value} (Count)")
```
